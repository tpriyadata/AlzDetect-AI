```python
import faiss
import numpy as np
import json
import os

print("Imports loaded!")
print(f"   FAISS version: {faiss.__version__}")
```

### Load Embeddings

```python
# Load Embeddings ────────────────
EMBEDDINGS_FILE = '../data/processed/embeddings.npy'
CHUNKS_FILE     = '../data/processed/chunks.json'

# Load embeddings
embeddings = np.load(EMBEDDINGS_FILE)
print(f" Embeddings loaded!")
print(f"   Shape: {embeddings.shape}")
print(f"   Type:  {embeddings.dtype}")

# Load chunks
with open(CHUNKS_FILE) as f:
    chunks = json.load(f)
print(f" Chunks loaded!")
print(f"   Total: {len(chunks):,}")

# Verify match
print(f"\n{'='*40}")
print(f"Embeddings: {embeddings.shape[0]:,}")
print(f"Chunks:     {len(chunks):,}")
if embeddings.shape[0] == len(chunks):
    print(f" MATCH! Ready to build index!")
else:
    print(f" MISMATCH! Check files!")
```

### Build FAISS Index

```python
# Convert to float32
embeddings = embeddings.astype('float32')

# Get dimensions
d = embeddings.shape[1]  # 768
print(f" Dimensions: {d}")

# Build index
print(f"  Building FAISS index...")
index = faiss.IndexFlatL2(d)
index.add(embeddings)

print(f"\n FAISS index built!")
print(f"   Total vectors: {index.ntotal:,}")
print(f"   Dimensions:    {d}")
```

### Save FIASS Index

```python
# Save FAISS Index ───────────────
import os

os.makedirs(
    '../vector_store/faiss_index',
    exist_ok=True
)

FAISS_FILE = '../vector_store/faiss_index/alzheimer.index'

faiss.write_index(index, FAISS_FILE)

size_mb = os.path.getsize(FAISS_FILE)/1024/1024

print(f"✅ FAISS index saved!")
print(f"   File:    {FAISS_FILE}")
print(f"   Vectors: {index.ntotal:,}")
print(f"   Size:    {size_mb:.1f} MB")
```
